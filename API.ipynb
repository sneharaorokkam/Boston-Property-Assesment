```python
import urllib.request
import xmltodict
import pandas as pd

# Starting URL
url = 'https://data.boston.gov/datastore/odata3.0/1000d81c-5bb5-49e8-a9ab-44cd042f1db2?$format=xml'

rows = []

i = 0
while url:
    i = i+1
    print("Loop -", i)
    # Sending request to the API
    fileobj = urllib.request.urlopen(url)

    # Decoding the response with UTF-8 encoding
    response_xml = fileobj.read().decode('utf-8')

    # Parsing the XML response into dictionary
    data_dict = xmltodict.parse(response_xml)

    # Extract the list of entries
    entries = data_dict['feed']['entry']

    # Each entry's data is under 'content' -> 'm:properties'
    # Extract that data and append it to the list of rows
    for entry in entries:
        properties = entry['content']['m:properties']
        row = {key: val.get('#text', None) for key, val in properties.items()}
        rows.append(row)

    # Reset URL to None
    url = None

    # Check for 'next' link
    links = data_dict['feed']['link']
    if isinstance(links, list):
        for link in links:
            if link['@rel'] == 'next':
                url = link['@href']
                break
    elif isinstance(links, dict) and links['@rel'] == 'next':
        url = links['@href']

# Convert the list of rows into a DataFrame
df = pd.DataFrame(rows)

```
